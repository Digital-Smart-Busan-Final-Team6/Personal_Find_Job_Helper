```python
post_improve_detail = toImproveDataFrame(post_detail)
```

```python
split_columns = ['main_tasks', 'requirements', 'preferred_points', 'benefits', 'hire_rounds']

for column in split_columns:
    post_improve_detail[column] = post_improve_detail[column].apply(
        lambda x: x.split('\n') if isinstance(x, str) else x)

post_improve_detail.drop(columns=['country'], inplace=True)
```

```python
post_improve_detail.drop_duplicates(subset=['position'], inplace=True)
```

```python
post_improve_detail.set_index('job_id', inplace=True)
```

```python
columns = post_improve_detail.columns
columns = columns.tolist()
for col in columns:
    post_improve_detail[col] = post_improve_detail[col].apply(
        lambda x: x.replace('\n', '') if isinstance(x, str) else x)
```

```python
post_improve_detail.rename(columns={
    'position': '제목',
    'intro': '회사 소개',
    'main_tasks': '주요 업무',
    'requirements': '자격 요건',
    'preferred_points': '우대 사항',
    'benefits': '복지',
    'hire_rounds': '채용절차',
    'attraction': '장점',
    'company': '회사명',
    'location': '근무지',
    'district': '지역',
    'category_parent': '직군',
    'category_child': '직무',
    'skill': '기술 스택',
    'annual_from': '요구 최소 경력',
    'annual_to': '요구 최대 경력',
    'is_newbie': '신입 지원 가능 여부',
    'due_time': '마감일',
}, inplace=True)
```

```python
import numpy as np

# 공백·탭·개행 모두 제거한 뒤 '' 이면 상시 공고
post_improve_detail["마감일"] = np.where(
    post_improve_detail["마감일"].astype(str).str.strip() == "",
    "상시 공고",
    post_improve_detail["마감일"]
)
```
