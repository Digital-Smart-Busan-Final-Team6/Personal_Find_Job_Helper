```python
from langchain_core.documents import Document

row_indices = sorted(data['회사명'].keys())

# 3) 행별로 dict로 재구성
records = [
    {field: data[field][i] for field in data}
    for i in row_indices
]

docs = []
for rec in records:
    company_name = rec.get('회사명', '')
    metadata = {'id': company_name, 'source': '회사 정보'}

    # page_content 직렬화
    lines = []
    for k, v in rec.items():
        if k == '회사명':
            continue
        if isinstance(v, list):
            # 리스트는 줄바꿈으로 합치기
            joined = "\n".join(v)
        else:
            joined = str(v)
        lines.append(f"{k}: {joined}")

    page_content = "\n".join(lines)
    docs.append(Document(page_content=page_content, metadata=metadata))
```

```python
# 캐시 존재 시 로드
from pathlib import Path
from typing import List
import json
from langchain_core.documents import Document

cache_dir = Path('Data_Files')
cache_file = cache_dir / '1000_50_docs_chunk'
if cache_file.exists():
    loaded: List[Document] = []
    with open(cache_file, 'r', encoding='utf-8') as f:
        for line in f:
            record = json.loads(line)
            loaded.append(Document(page_content=record["page_content"], metadata=record["metadata"]))
    print(f"캐시에서 {len(loaded)}개의 문서 청크 로드 완료")
```

```python
from Run_Pipeline.Embedding_DB import EmbeddingDB
from Run_Pipeline.Document_Splitter import DocumentSplitter
import os
from pathlib import Path
from Run_Pipeline.Document_Loader import DocumentLoader
from Run_Pipeline.Env_Loader import EnvLoader

EnvLoader.load_local()

BASE_DIR = Path(os.getcwd())

kind = "json"
file_path = BASE_DIR / "Data_Files"
chunk_size = 1000
overlap_size = 50
persist_dir = BASE_DIR / "Data_Files" / f"{kind}_{chunk_size}"
device = "mps"
retriever_mode = 1
k = 3
engine_num = 1
backend_num = 1

loader = DocumentLoader(file_path, kind)  # 일반 파일들을 Document형태로 변환해서 다 불러옴
docs = loader._load_json_files()

# ④ 문서 분할
splitter = DocumentSplitter(chunk_size=chunk_size, overlap=overlap_size)  # 이미 분할된 문서가 있다면 로드하고 아니면 새로 스플릿함
chunks = splitter.split(docs, cache_dir=file_path)

# ⑤ 임베딩 & DB 생성/로드
embed_db = EmbeddingDB(model_name="nlpai-lab/KURE-v1", device=device,
                       persist_dir=persist_dir)  # 이미 임베딩된 DB가 있다면 로드하고 아니면 새로 임베딩함
db = embed_db.get_or_create_db(chunks)
```
