```python
# df 전체 에서 "더보기"라는 글자 제거

df_cleaned = df.applymap(lambda x: x.replace("더보기", "") if isinstance(x, str) else x)
```

```python
# df의 회사 소개 안에 리스트 안의 글자에서 더보기라는 글자 제거
df_cleaned['회사 소개'] = df_cleaned['회사 소개'].apply(
    lambda x: [item.replace("더보기", "") for item in x] if isinstance(x, list) else x)
```

```python
from langchain_core.documents import Document

row_indices = sorted(data['회사명'].keys())

# 3) 행별로 dict로 재구성
records = [
    { field: data[field][i] for field in data }
    for i in row_indices
]


docs = []
for rec in records:
    company_name = rec.get('회사명', '')
    metadata = {'id': company_name, 'source': '회사 정보'}

    # page_content 직렬화
    lines = []
    for k, v in rec.items():
        if k == '회사명':
            continue
        if isinstance(v, list):
            # 리스트는 줄바꿈으로 합치기
            joined = "\n".join(v)
        else:
            joined = str(v)
        lines.append(f"{k}: {joined}")

    page_content = "\n".join(lines)
    docs.append(Document(page_content=page_content, metadata=metadata))
```
